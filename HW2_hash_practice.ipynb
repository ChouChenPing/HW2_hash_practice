```python
path = '/content/hw2_data.txt' #老師您好，我是用google colab進行程式撰寫，因此這邊的空格可能需要填入上傳檔案的路徑位置
food = open(path, 'r')

food_name={}
for line in food.readlines():
  line = line.split('\n')[0]
  food_name[line]=food_name.get(line, 0)+1 #get function本身就可以在dictionary裡新增一組key-value，但是又只會回傳value
print('總共有這些食物:',sorted(food_name.keys()))
print('關於食物計數的方面:',food_name)
```

```python
import matplotlib.pyplot as plt
import numpy as np

bigtosmall = sorted(food_name.items(), key=lambda x:x[1], reverse=True)
food_names = [item[0] for item in bigtosmall]
food_counts = [item[1] for item in bigtosmall]
plt.bar(food_names, food_counts)
plt.show()

```
